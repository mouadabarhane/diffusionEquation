```python
import numpy as np
import matplotlib.pyplot as plt

# Paramtres du problme
L = 1.0  # Longueur spatiale
T = 1.0  # Temps total
Nx = 100  # Nombre de divisions spatiales
Nt = 1000  # Nombre de divisions temporelles
D = 0.1  # Coefficient de diffusion

dx = L / Nx  # Pas d'espace
dt = T / Nt  # Pas de temps
alpha = D * dt / (dx ** 2)  # Paramtre alpha

# Conditions initiales et aux bords
u = np.zeros(Nx+1)  # Initialise u  zro partout
u[int(0.5 / dx):int(0.6 / dx + 1)] = 1.0  # Condition initiale : pulse
u[0] = 0.0  # Condition de Dirichlet au bord gauche
u[Nx] = 0.0  # Condition de Dirichlet au bord droit

# Rsolution de l'quation de diffusion avec la mthode des volumes finis
for t in range(Nt):
    unew = u.copy()  # Copie de u pour la mise  jour
    for i in range(1, Nx):  # Boucle sur les points intrieurs
        u[i] = unew[i] + alpha * (unew[i - 1] - 2 * unew[i] + unew[i + 1])
    u[0] = 0.0  # Condition de Dirichlet au bord gauche
    u[Nx] = 0.0  # Condition de Dirichlet au bord droit

    # Trac de la solution  certaines tapes temporelles spcifiques
    if t % 100 == 0:  # Tracer tous les 100 pas de temps
        x = np.linspace(0, L, Nx+1)  # Cre une grille spatiale
        plt.plot(x, u, label=f'Temps {t}')

# Affichage de la lgende et des dtails du graphique
plt.title('Solution de l quation de diffusion avec la mthode des volumes finis')
plt.xlabel('Position')
plt.ylabel('Valeur de u')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Paramtres du problme
Lx = 1.0  # Longueur spatiale en x
Ly = 1.0  # Longueur spatiale en y
T = 1.0  # Temps total
Nx = 50  # Nombre de divisions spatiales en x
Ny = 50  # Nombre de divisions spatiales en y
Nt = 1000  # Nombre de divisions temporelles
D = 0.1  # Coefficient de diffusion

dx = Lx / Nx  # Pas d'espace en x
dy = Ly / Ny  # Pas d'espace en y
dt = T / Nt  # Pas de temps
alpha_x = D * dt / (dx ** 2)  # Paramtre alpha en x
alpha_y = D * dt / (dy ** 2)  # Paramtre alpha en y

# Conditions initiales et aux bords
u = np.zeros((Nx+1, Ny+1))  # Initialise u  zro partout
u[int(0.4 / dx):int(0.6 / dx + 1), int(0.4 / dy):int(0.6 / dy + 1)] = 1.0  # Condition initiale : pulse
u[:, 0] = 0.0  # Condition de Dirichlet au bord infrieur
u[:, Ny] = 0.0  # Condition de Dirichlet au bord suprieur
u[0, :] = 0.0  # Condition de Dirichlet au bord gauche
u[Nx, :] = 0.0  # Condition de Dirichlet au bord droit

# Rsolution de l'quation de diffusion en 2D avec la mthode des diffrences finies
for t in range(Nt):
    unew = u.copy()  # Copie de u pour la mise  jour
    for i in range(1, Nx):  # Boucle sur les points intrieurs en x
        for j in range(1, Ny):  # Boucle sur les points intrieurs en y
            u[i, j] = unew[i, j] + alpha_x * (unew[i - 1, j] - 2 * unew[i, j] + unew[i + 1, j]) \
                + alpha_y * (unew[i, j - 1] - 2 * unew[i, j] + unew[i, j + 1])
    # Conditions de Dirichlet aux bords
    u[:, 0] = 0.0
    u[:, Ny] = 0.0
    u[0, :] = 0.0
    u[Nx, :] = 0.0

    # Trac de la solution  certaines tapes temporelles spcifiques
    if t % 100 == 0:  # Tracer tous les 100 pas de temps
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        x = np.linspace(0, Lx, Nx+1)
        y = np.linspace(0, Ly, Ny+1)
        X, Y = np.meshgrid(x, y)
        ax.plot_surface(X, Y, u, cmap='viridis')
        ax.set_title(f'Temps {t}')
        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Valeur de u')
        plt.show()
```
